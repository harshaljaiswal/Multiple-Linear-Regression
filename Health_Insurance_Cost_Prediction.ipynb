```python
#importing libraries
import pandas as pd
import numpy as np

# Loading the dataset
data = pd.read_csv("insurance.csv")
x= data.iloc[:,:-1].values
y=data.iloc[:,6].values

# dealing with categorical data
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder = LabelEncoder()
x[:,1] = labelencoder.fit_transform(x[:,1])
x[:,4] = labelencoder.fit_transform(x[:,4])
x[:,5] = labelencoder.fit_transform(x[:,5])
onehotencoder =OneHotEncoder(categorical_features=[1,4,5])
x = onehotencoder.fit_transform(x).toarray()
x=x[:,[1,3,5,6,7,8,9,10]]

# Partitioning data
from sklearn.cross_validation import train_test_split
x_train,x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 0)

#Training the MLR
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)

#Testing Built MLR
y_prediction = regressor.predict(x_test)
```

```python
#CHOOSING OPTIMAL VARIABLES WITH BACKWARD ELIMINATION

import statsmodels.formula.api as sm
x = np.append(arr= np.ones((1338,1)).astype(int), values=x, axis=1)

#1ST ITERATION
x_opt = x[:,[0,1,2,3,4,5,6,7,8]]
regressor_OLS = sm.OLS(endog=y, exog=x_opt).fit()
regressor_OLS.summary()
```

```python
#2ND ITERATION: AFTER REMOVING VARIABLE X1
x_opt = x[:,[0,2,3,4,5,6,7,8]]
regressor_OLS = sm.OLS(endog=y, exog=x_opt).fit()
regressor_OLS.summary()
```

```python
#3RD ITERATION: AFTER REMOVING VARIABLE X2
x_opt = x[:,[0,2,4,5,6,7,8]]
regressor_OLS = sm.OLS(endog=y, exog=x_opt).fit()
regressor_OLS.summary()
```

```python
#4TH ITERATION: AFTER REMOVING VARIABLE X3 
x_opt = x[:,[0,2,4,6,7,8]]
regressor_OLS = sm.OLS(endog=y, exog=x_opt).fit()
regressor_OLS.summary()
```

```python
#5TH ITERATION : AFTER REMOVING VARIABLE X2
x_opt = x[:,[0,2,6,7,8]]
regressor_OLS = sm.OLS(endog=y, exog=x_opt).fit()
regressor_OLS.summary()
```
